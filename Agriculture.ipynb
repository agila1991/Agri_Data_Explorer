```python
def create_table(df, table_name):
    dtype_mapping = {
        "int64": "BIGINT",
        "float64": "DOUBLE",
        "object": "TEXT",
        "bool": "BOOLEAN",
        "datetime64[ns]": "DATETIME"
    }
    sql_columns = []
    for col in df.columns:
        col_name = col.strip().upper().replace(" ", "_").replace("(", "").replace(")", "").replace("-", "_")
        dtype = str(df[col].dtype)
        mysql_type = dtype_mapping.get(dtype, "TEXT")
        sql_columns.append(f"`{col_name}` {mysql_type}")

    columns_block = ",\n  ".join(sql_columns)
    create_table_sql = f"CREATE TABLE `{table_name}` (\n  {columns_block}\n);"
    return create_table_sql

sql_statement = create_table(df, "Agriculture_data")

mycursor.execute(sql_statement)
mydb.commit()
```

```python
mycursor.execute('''SELECT 
    STATE_NAME, 
    SUM(RICE_PRODUCTION_1000_TONS) AS TOTAL_PRODUCTION
FROM 
    agriculture_data
GROUP BY 
    STATE_NAME
ORDER BY 
    TOTAL_PRODUCTION DESC
LIMIT 3;'''
    )
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],tablefmt='psql'))
```

```python
mycursor.execute('''WITH last5years AS (
    SELECT DISTINCT YEAR
    FROM agriculture_data
    ORDER BY YEAR DESC
    LIMIT 5
),
filtered_data AS (
    SELECT *
    FROM agriculture_data
    WHERE YEAR IN (SELECT YEAR FROM last5years)
),
yield_change AS (
    SELECT 
        DIST_NAME,
        MAX(YEAR) AS LATEST_YEAR,
        MIN(YEAR) AS EARLIEST_YEAR,
        MAX(WHEAT_YIELD_KG_PER_HA) AS MAX_YIELD,
        MIN(WHEAT_YIELD_KG_PER_HA) AS MIN_YIELD,
        (MAX(WHEAT_YIELD_KG_PER_HA) - MIN(WHEAT_YIELD_KG_PER_HA)) AS YIELD_INCREASE
    FROM filtered_data
    GROUP BY DIST_NAME
)
SELECT 
    DIST_NAME,
    YIELD_INCREASE
FROM yield_change
ORDER BY YIELD_INCREASE DESC
LIMIT 5;
'''
    )
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],tablefmt='psql'))
```

```python
mycursor.execute('''
    WITH last5years AS (
    SELECT DISTINCT YEAR
    FROM agriculture_data
    ORDER BY YEAR DESC
    LIMIT 5
),
filtered_data AS (
    SELECT *
    FROM agriculture_data
    WHERE YEAR IN (SELECT YEAR FROM last5years)
),
state_yearly_production AS (
    SELECT 
        STATE_NAME,
        YEAR,
        SUM(OILSEEDS_PRODUCTION_1000_TONS) AS TOTAL_PRODUCTION
    FROM filtered_data
    GROUP BY STATE_NAME, YEAR
),
state_growth AS (
    SELECT 
        STATE_NAME,
        MAX(YEAR) AS LATEST_YEAR,
        MIN(YEAR) AS EARLIEST_YEAR,
        MAX(TOTAL_PRODUCTION) AS MAX_PROD,
        MIN(TOTAL_PRODUCTION) AS MIN_PROD,
        ((MAX(TOTAL_PRODUCTION) - MIN(TOTAL_PRODUCTION)) / NULLIF(MIN(TOTAL_PRODUCTION), 0)) * 100 AS GROWTH_RATE_PERCENT
    FROM state_yearly_production
    GROUP BY STATE_NAME
)
SELECT 
    STATE_NAME,
    GROWTH_RATE_PERCENT
FROM state_growth
ORDER BY GROWTH_RATE_PERCENT DESC
LIMIT 3;
'''
    )
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],tablefmt='psql'))
```

```python
mycursor.execute('''
SELECT
    DIST_NAME,
    (
        SUM((RICE_AREA_1000_HA - area_avg) * (RICE_PRODUCTION_1000_TONS - prod_avg)) /
        (SQRT(SUM(POW(RICE_AREA_1000_HA - area_avg, 2))) * SQRT(SUM(POW(RICE_PRODUCTION_1000_TONS - prod_avg, 2))))
    ) AS rice_corr
FROM (
    SELECT 
        DIST_NAME,
        RICE_AREA_1000_HA,
        RICE_PRODUCTION_1000_TONS,
        AVG(RICE_AREA_1000_HA) OVER (PARTITION BY DIST_NAME) AS area_avg,
        AVG(RICE_PRODUCTION_1000_TONS) OVER (PARTITION BY DIST_NAME) AS prod_avg
    FROM agriculture_data
    WHERE RICE_AREA_1000_HA IS NOT NULL AND RICE_PRODUCTION_1000_TONS IS NOT NULL
) AS subquery
GROUP BY DIST_NAME
ORDER BY rice_corr DESC;

'''
    )
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],tablefmt='psql'))
```

```python
mycursor.execute('''
WITH top_states AS (
  SELECT STATE_NAME
  FROM agriculture_data
  GROUP BY STATE_NAME
  ORDER BY SUM(COTTON_PRODUCTION_1000_TONS) DESC
  LIMIT 5
)
SELECT YEAR, STATE_NAME, SUM(COTTON_PRODUCTION_1000_TONS) AS TOTAL_COTTON_PRODUCTION
FROM agriculture_data
WHERE STATE_NAME IN (SELECT STATE_NAME FROM top_states)
GROUP BY YEAR, STATE_NAME
ORDER BY YEAR, STATE_NAME;
'''
    )
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],tablefmt='psql'))
```

```python
mycursor.execute('''
WITH top_states AS (
  SELECT STATE_NAME
  FROM agriculture_data
  GROUP BY STATE_NAME
  ORDER BY SUM(WHEAT_PRODUCTION_1000_TONS + RICE_PRODUCTION_1000_TONS) DESC
  LIMIT 5
)
SELECT YEAR, STATE_NAME,
       SUM(RICE_PRODUCTION_1000_TONS) AS RICE_PRODUCTION,
       SUM(WHEAT_PRODUCTION_1000_TONS) AS WHEAT_PRODUCTION
FROM agriculture_data
WHERE STATE_NAME IN (SELECT STATE_NAME FROM top_states)
  AND YEAR >= YEAR(CURDATE()) - 10
GROUP BY YEAR, STATE_NAME
ORDER BY STATE_NAME, YEAR;

'''
    )
out=mycursor.fetchall()
print(tabulate(out,headers=[i[0] for i in mycursor.description],tablefmt='psql'))
```
